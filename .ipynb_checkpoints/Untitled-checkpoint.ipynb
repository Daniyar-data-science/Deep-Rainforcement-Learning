```python
import re
import random
import torch
import torch.nn as nn
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

PAD_TOKEN = 0
SOS_TOKEN = 1
EOS_TOKEN = 2
UNK_TOKEN = 3

MAX_LENGTH = 10
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z0-9?.!,]+", " ", text)
    return text.strip()


class Vocab:
    def __init__(self):
        self.word2index = {
            "<PAD>": PAD_TOKEN,
            "<SOS>": SOS_TOKEN,
            "<EOS>": EOS_TOKEN,
            "<UNK>": UNK_TOKEN
        }
        self.index2word = {v: k for k, v in self.word2index.items()}
        self.word_count = {}
        self.n_words = 4

    def add_sentence(self, sentence):
        for word in sentence.split():
            self.add_word(word)

    def add_word(self, word):
        if word not in self.word2index:
            self.word2index[word] = self.n_words
            self.index2word[self.n_words] = word
            self.word_count[word] = 1
            self.n_words += 1
        else:
            self.word_count[word] += 1

    def sentence_to_indexes(self, sentence):
        return [
            self.word2index.get(word, UNK_TOKEN)
            for word in sentence.split()
        ] + [EOS_TOKEN]


def pad_sequence(seq, max_length=MAX_LENGTH):
    seq = seq[:max_length]
    return seq + [PAD_TOKEN] * (max_length - len(seq))
```

```python
pairs = [
    ["hi", "hello"],
    ["hello", "hi there"],
    ["how are you", "i am fine"],
    ["i am fine", "that is good"],
    ["what is your name", "i am a bot"],
    ["nice to meet you", "nice to meet you too"],
    ["where are you from", "i am from the internet"],
    ["what are you doing", "i am talking to you"],
    ["do you like music", "yes i like music"],
    ["what kind of music", "i like all kinds"],
    ["bye", "goodbye"],
    ["goodbye", "see you later"]
]

vocab = Vocab()

for p in pairs:
    vocab.add_sentence(clean_text(p[0]))
    vocab.add_sentence(clean_text(p[1]))
```

```python
class EncoderRNN(nn.Module):
    def __init__(self, input_size, embed_size, hidden_size, num_layers=2):
        super().__init__()
        self.embedding = nn.Embedding(input_size, embed_size, padding_idx=PAD_TOKEN)
        self.lstm = nn.LSTM(embed_size, hidden_size, num_layers=num_layers, batch_first=True)

    def forward(self, x):
        embedded = self.embedding(x)
        _, (hidden, cell) = self.lstm(embedded)
        return hidden, cell


class DecoderRNN(nn.Module):
    def __init__(self, output_size, embed_size, hidden_size, num_layers=2):
        super().__init__()
        self.embedding = nn.Embedding(output_size, embed_size, padding_idx=PAD_TOKEN)
        self.lstm = nn.LSTM(embed_size, hidden_size, num_layers=num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, hidden, cell):
        x = x.unsqueeze(1)
        embedded = self.embedding(x)
        output, (hidden, cell) = self.lstm(embedded, (hidden, cell))
        prediction = self.fc(output.squeeze(1))
        return prediction, hidden, cell
```

```python
INPUT_SIZE = vocab.n_words
OUTPUT_SIZE = vocab.n_words
EMBED_SIZE = 128
HIDDEN_SIZE = 256

encoder = EncoderRNN(INPUT_SIZE, EMBED_SIZE, HIDDEN_SIZE).to(device)
decoder = DecoderRNN(OUTPUT_SIZE, EMBED_SIZE, HIDDEN_SIZE).to(device)

optimizer = optim.Adam(list(encoder.parameters()) + list(decoder.parameters()))
criterion = nn.CrossEntropyLoss(ignore_index=PAD_TOKEN)
```

```python
def save_model():
    torch.save({
        'encoder': encoder.state_dict(),
        'decoder': decoder.state_dict(),
        'optimizer': optimizer.state_dict(),
        'vocab': vocab.__dict__
    }, "chatbot.pth")
    print("✅ Model saved")


def load_model():
    global vocab
    try:
        checkpoint = torch.load("chatbot.pth", map_location=device)

        encoder.load_state_dict(checkpoint['encoder'])
        decoder.load_state_dict(checkpoint['decoder'])
        optimizer.load_state_dict(checkpoint['optimizer'])

        vocab.__dict__ = checkpoint['vocab']

        print("✅ Model loaded")
    except:
        print("⚠️ No saved model found")
```

```python
def train_step(input_tensor, target_tensor):
    encoder.train()
    decoder.train()

    optimizer.zero_grad()

    hidden, cell = encoder(input_tensor)
    decoder_input = torch.tensor([SOS_TOKEN], device=device)

    loss = 0

    for t in range(target_tensor.size(1)):
        output, hidden, cell = decoder(decoder_input, hidden, cell)
        loss += criterion(output, target_tensor[:, t])

        teacher_force = random.random() < 0.5
        top1 = output.argmax(1)

        decoder_input = target_tensor[:, t] if teacher_force else top1

    loss.backward()
    optimizer.step()

    return loss.item()
```

```python
load_model()

for epoch in range(100):
    total_loss = 0
    for p in pairs:
        input_indexes = pad_sequence(vocab.sentence_to_indexes(clean_text(p[0])))
        target_indexes = pad_sequence(vocab.sentence_to_indexes(clean_text(p[1])))

        input_tensor = torch.tensor([input_indexes], device=device)
        target_tensor = torch.tensor([target_indexes], device=device)

        loss = train_step(input_tensor, target_tensor)
        total_loss += loss

    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss:.4f}")

save_model()
```

```python
def greedy_decode(sentence):
    encoder.eval()
    decoder.eval()

    indexes = pad_sequence(vocab.sentence_to_indexes(clean_text(sentence)))
    input_tensor = torch.tensor([indexes], device=device)

    with torch.no_grad():
        hidden, cell = encoder(input_tensor)

        decoder_input = torch.tensor([SOS_TOKEN], device=device)
        result = []

        for _ in range(MAX_LENGTH):
            output, hidden, cell = decoder(decoder_input, hidden, cell)
            top1 = output.argmax(1).item()

            if top1 == EOS_TOKEN:
                break

            result.append(vocab.index2word.get(top1, "<UNK>"))
            decoder_input = torch.tensor([top1], device=device)

    return " ".join(result)


def predict_next_word(sentence):
    encoder.eval()
    decoder.eval()

    indexes = pad_sequence(vocab.sentence_to_indexes(clean_text(sentence)))
    input_tensor = torch.tensor([indexes], device=device)

    with torch.no_grad():
        hidden, cell = encoder(input_tensor)
        decoder_input = torch.tensor([SOS_TOKEN], device=device)

        output, _, _ = decoder(decoder_input, hidden, cell)
        top1 = output.argmax(1).item()

    return vocab.index2word.get(top1, "<UNK>")
```

```python
def learn_from_user(user_input, bot_response):
    vocab.add_sentence(clean_text(user_input))
    vocab.add_sentence(clean_text(bot_response))

    input_indexes = pad_sequence(vocab.sentence_to_indexes(clean_text(user_input)))
    target_indexes = pad_sequence(vocab.sentence_to_indexes(clean_text(bot_response)))

    input_tensor = torch.tensor([input_indexes], device=device)
    target_tensor = torch.tensor([target_indexes], device=device)

    loss = train_step(input_tensor, target_tensor)
    return loss
```

```python
def chat_once():
    user_input = input("You: ")

    suggestion = predict_next_word(user_input)
    print(f"Suggestion: {user_input} {suggestion}")

    response = greedy_decode(user_input)
    print(f"Bot: {response}")

    loss = learn_from_user(user_input, response)
    print(f"📈 Learned (loss={loss:.4f})")

    save_model()
```

```python
chat_once()
chat_once()
chat_once()
```
